# Idea
Solve heat equation for the glacier surface temperature dynamics:
\begin{equation}
    \dfrac{\partial T}{\partial t} = \alpha \nabla^2 T = \alpha \left(\dfrac{\partial^2 T}{\partial x^2} + \dfrac{\partial^2 T}{\partial y^2}\right).
\end{equation}


```python
from heat import predict, show
```

```python
image_path = 'data/landsat_7_img_0_snow_ice.npy'
alpha = 10e-7 # Thermal diffusivity of snow and ice ranges between 9e-7 to 15e-7 in m^2 s^{-1} 
days = 30 # Number of days to predict
```

```python
data = predict(image_path, alpha, days)
```

```python
show(data)
```
